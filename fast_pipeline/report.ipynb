```python
from statistics import median, mean


def add_logs(name: str, times: list, time_logs: dict):
    time_logs["name"].append(name)
    time_logs["min"].append(min(times))
    time_logs["max"].append(max(times))
    time_logs["median"].append(median(times))
    time_logs["mean"].append(mean(times))


time_logs = {"name": [], "min": [], "max": [], "mean": [], "median": []}
```

```python
from task2.run_epoch import run_epoch, DataMode


times = run_epoch(DataMode.BRAIN, "wiki.train.raw")
add_logs("BrainDataset", times, time_logs)
```

```python
times = run_epoch(DataMode.BIG_BRAIN, "wiki.train.raw")
add_logs("BigBrainDataset", times, time_logs)
```

```python
for k in [1, 5, 10, 20, 50, 640]:
    times = run_epoch(DataMode.ULTRA_DUPER_BIG_BRAIN, "wiki.train.raw", k)
    add_logs(f"UltraDuperBigBrainDataset_k={k}", times, time_logs)
```
